```python
import pandas as pd

data = {
    "patient_id": [1,2,3,4,5,6,7,8],
    "notes": [
        "patient has diabetes and high glucose",
        "no signs of diabetes",
        "blood sugar elevated",
        "patient using insulin daily",
        "normal condition",
        "glucose levels high and diabetes suspected",
        "insulin prescribed with high glucose",
        "routine checkup no issues"
    ],
    "num_codes": [5, 0, 3, 6, 0, 4, 7, 0]
}

df = pd.DataFrame(data)
print(df)

keywords = ["diabetes", "glucose", "insulin", "blood sugar"]

print(keywords)

def keyword_score(text):
    return sum(word in text.lower() for word in keywords)

df["keyword_score"] = df["notes"].apply(keyword_score)

anchors = ["insulin"]

def anchor_flag(text):
    return int(any(word in text.lower() for word in anchors))

df["anchor"] = df["notes"].apply(anchor_flag)

import numpy as np

df["probability"] = (df["num_codes"] / df["num_codes"].max()) * 0.5 + df["keyword_score"] * 0.1 + df["anchor"] * 0.4

df["label"] = df["probability"].apply(lambda x: 1 if x > 0.5 else 0)

df["text_length"] = df["notes"].apply(len)

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = df[["keyword_score", "anchor", "num_codes", "text_length"]]
y = df["label"]

X_train, X_test, y_train, y_test = train_test_split(X, y)

model = RandomForestClassifier()
model.fit(X_train, y_train)

import pandas as pd

importance = pd.Series(model.feature_importances_, index=X.columns)
print(importance)

```
